```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("computadoras.csv")
df
```

# • Carga el archivo CSV y muestra las primeras y últimas filas del dataset.

```python
df.head()
df.tail()
```

###• Genera una tabla que muestre cuántas computadoras de cada marca hay en el dataset.

```python
computadoras_por_marca = df['Marca'].value_counts().reset_index()
computadoras_por_marca.columns = ['Marca', 'Cantidad']
print(computadoras_por_marca)
```

# • Extrae y muestra solo las computadoras cuyo estado sea "Usado".

```python
computadoras_usadas = df[df['Estado'] == 'Usado']
print(computadoras_usadas)
```

# Encuentra y muestra el registro de la computadora más cara en el dataset.

```python
computadora_mas_cara = df[df['Precio'] == df['Precio'].max()]
print (computadora_mas_cara)
```

# Añade una columna al dataset que indique el precio por GB de almacenamiento.

```python
df['Precio_por_GB'] = df['Precio'] / df['Almacenamiento (GB)']
print(df.head())
```

# Gráficos
### Gráfico de barras: Promedio de precios por marca
### • Representa el precio promedio de las computadoras para cada marca.

```python
promedio_precios_marca = df.groupby('Marca')['Precio'].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 6))  
promedio_precios_marca.plot(kind='bar', color='skyblue')

plt.title('Promedio de Precios por Marca')
plt.xlabel('Marca')
plt.ylabel('Precio Promedio')
plt.xticks(rotation=45) 
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

# Gráfico de líneas: Tendencia de precios promedio por año
# • Muestra cómo cambia el precio promedio de las computadoras a lo largo de los años.

```python
promedio_precios_anio = df.groupby('Año')['Precio'].mean().sort_index()

plt.figure(figsize=(12, 6)) 
plt.plot(promedio_precios_anio.index, promedio_precios_anio.values, marker='o', linestyle='-', color='green')

plt.title('Tendencia de Precios Promedio de Computadoras por Año')
plt.xlabel('Año')
plt.ylabel('Precio Promedio')
plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

# Gráfico circular: Proporción de estados de computadoras
# • Genera un gráfico de pastel que represente la proporción de computadoras nuevas, usadas y certificadas.

```python
proporcion_estados = df['Estado'].value_counts()

plt.figure(figsize=(8, 8))  
plt.pie(proporcion_estados, labels=proporcion_estados.index, autopct='%1.1f%%', startangle=140, colors=['lightgreen', 'skyblue', 'lightcoral'])

plt.title('Proporción de Estados de Computadoras')
plt.axis('equal')  
plt.show()
```

# Diagrama de dispersión: Relación entre precio y almacenamiento
# • Crea un scatter plot que muestre la relación entre el precio y la capacidad de almacenamiento.


```python
plt.figure(figsize=(10, 6)) 
plt.scatter(df['Almacenamiento (GB)'], df['Precio'], alpha=0.6, color='purple')

plt.title('Relación entre Precio y Almacenamiento')
plt.xlabel('Almacenamiento (GB)')
plt.ylabel('Precio')
plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

# Histograma: Distribución de precios de computadoras usadas
# • Representa la distribución de precios de las computadoras cuyo estado es "Usado".

```python
computadoras_usadas = df[df['Estado'] == 'Usado']

plt.figure(figsize=(10, 6)) 
plt.hist(computadoras_usadas['Precio'], bins=20, color='orange', edgecolor='black')

plt.title('Distribución de Precios de Computadoras Usadas')
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
